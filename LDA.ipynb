```python
# necessary import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_covtype
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, confusion_matrix,f1_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
cov_type = fetch_covtype()
dataset = pd.DataFrame(columns=cov_type.feature_names,data=cov_type.data)
dataset['target'] = cov_type.target
dataset

```

```python
# divide the dataset into class and target variable
X = dataset.iloc[:, 0:53].values
y = dataset.iloc[:, 54].values

# Preprocess the dataset and divide into train and test
sc = StandardScaler()
X = sc.fit_transform(X)
le = LabelEncoder()
y = le.fit_transform(y)
X_train, X_test,y_train, y_test = train_test_split(X, y,test_size=0.2)




```

```python
# apply Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(n_components=6)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

classifiers = [RandomForestClassifier(max_depth=2,random_state=0),
               KNeighborsClassifier(n_neighbors=4),
               DecisionTreeClassifier(random_state=42),
               GaussianNB()]
classifiers_names = ["RandomForestClassifier",
               "KNeighborsClassifier",
               "DecisionTreeClassifier",
               "GaussianNB"]
classifiers_scores = []



for index,classifier in enumerate(classifiers):
   
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    
    accuracy_score_v = accuracy_score(y_test, y_pred)
    f1_score_v = f1_score(y_test, y_pred,average="weighted")
    # print the accuracy and confusion matrix
    print(f"        {classifiers_names[index]}")
    print('Accuracy : ' + str(accuracy_score_v))
    print('F1 score : ' + str(f1_score_v))
    conf_m = confusion_matrix(y_test, y_pred)
    print(conf_m)
    print("-------------------------------------------------")


classifiers_scores.sort()
```
